# Understanding Image

In this section we are going to import an Image and then understand the nature in which the image is represented in the Digital Form. We are going to use the Numpy and Matplotlib Library Since they have image handling capacity.

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt #imread, imsave, imshow
```

Reading the Image File

```python
colour_image = plt.imread("Image-test/PNG/Image01.png")
```

Displaying the Image

```python
plt.imshow(colour_image)
```

Finding the Data type in Which the Image is Loaded.

```python
type(colour_image)
```

Since the Image is Stored in N-Dimensional Array (or ndarray) we are going to find the dimensions of the Image.

```python
colour_image.shape
```

We Know that the dimension of the Image is 600x800 pixels. So the colour information is stored in the Fields corresponding to the colour of the image in different pixels.

Now we will Find out which Channel out of the 4 Has the Information of What Colour.

```python
#Keeping the first Channel And Making others zero
plt.subplot(2, 2, 1)
x = colour_image.copy()
x[:,:,1] = 0*x[:,:,1]
x[:,:,2] = 0*x[:,:,2]
x[:,:,3] = 1
plt.axis('off')
plt.title("Channel 0")
plt.imshow(x)

#Keeping the second Channel And Making others zero
plt.subplot(2, 2, 2)
x = colour_image.copy()
x[:,:,0] = 0*x[:,:,0]
x[:,:,2] = 0*x[:,:,2]
x[:,:,3] = 1
plt.axis('off')
plt.title("Channel 1")
plt.imshow(x)

#Keeping the third Channel And Making others zero
plt.subplot(2, 2, 3)
x = colour_image.copy()
x[:,:,0] = 0*x[:,:,0]
x[:,:,1] = 0*x[:,:,1]
x[:,:,3] = 1
plt.axis('off')
plt.title("Channel 2")
plt.imshow(x)

#Keeping the fourth Channel And Making others zero
plt.subplot(2, 2, 4)
x = colour_image.copy()
x[:,:,0] = 0*x[:,:,0]
x[:,:,1] = 0*x[:,:,1]
x[:,:,2] = 0*x[:,:,2]
plt.axis('off')
plt.title("Channel 3")
plt.imshow(x)
```

Therefore From the Above Outputs We can conclude that the above Image format contains: <br>
Red at Channel 0<br>
Green at Channel 1<br>
Blue at Channel 2<br>
Transparency at Channel 3

### The Method to Convert RGB to Gray-scale image

```python
def RGB_to_Gray(X):
    '''
      Argument(s):
    ----------------
    X    :    Is an ndarray containing a Colour Image
    
      Returns:
    ----------------
    YIQ : an ndarray containing the Grayscale Image
    '''
    x, y, _ = X.shape
    shape = (x,y) 
    YIQ = np.zeros(shape, dtype=X.dtype)
    YIQ[:,:] = 0.299*X[:,:,0] + 0.587*X[:,:,1] + 0.114*X[:,:,2]
    return YIQ
```

The Method Call

```python
gray_image = RGB_to_Gray(colour_image)
```

```python
gray_image.shape
```

```python
plt.imshow(gray_image, cmap='gray')
```

```python
plt.imsave("test.png", gray_image, cmap='gray')
```
